```python
import skimage
import numpy as np  

def padding_img(img):
        w,h = img.shape
        h1 = (h//64 + 1)*64
        tmp = np.zeros([h1,h1])
        x_start = int((h1 -w)//2)
        y_start = int((h1 -h)//2)
        tmp[x_start:x_start+w,y_start:y_start+h] = img
        return tmp

img = skimage.io.imread("E:\Key_File\硬币数据集\y+\y_0.bmp")
img = skimage.transform.resize(img,(512,512))
img = padding_img(img)
img = np.expand_dims(img,0)

img = np.expand_dims(img,0)
img.shape
```
